```python
from datetime import datetime

class Point:
    # Constructor for the Point class. It has latitude and longitude defined in the class.
    def __init__(self, lat, lon, tp):
        self.latitude = lat
        self.longitude = lon
        self.type = tp
        
    def coordinates(self):
        return [self.longitude, self.latitude]
    
    

#Lists are created to append the respective objects into it, that are read from the file.
stn = []
bikes_avail = []
total_docks = []
docks_serv = []
docks_avail = []
stats = []
time_stp = []
statid = []
    

# The BikeStation class is declared with a minimum of 6 attributes defined in the init function.
class BikeStation:
    def __init__(self, id, timestamp, station_name, available_docks, available_bikes, status):
        self.id = id
        self.timestamp = timestamp
        self.stn = station_name
        self.avdocks = available_docks
        self.avbikes = available_bikes
        self.status = status
   
#Status is made private by declaring the property decorator.
    @property
    def status(self):
        return self.__status
    
    @status.setter
    def status(self, a):
        a = str(a)
        if len(a) > 0:
            self.__status = a
  
#A calculated method is created to find the percentage_full.
    def calc_percent_full(self):
        per = (self.avbikes / self.avdocks) * 100
        return round(per)
    
    def calc_docks_in_service(self):
        return (self.avdocks + self.avbikes)
 
#This method is print the last line of the expected output.
    def present_avail_bikes(self):
        #print(str(list(self.id)))
        print("Stations: " + str(list(map(int, self.id))) + " Bikes Available " + str(sum(self.avbikes)) + " Docks Available " + str(sum(self.avdocks)))

#Map function used here is not taught. This is used to convert the station id numbers that are in string to int and still
#keep them intact in list.
    
    def __str__(self):
        for i in range(0, len(self.stn)): 
            print(self.stn[i] + ' had' + ' ' + str(self.avbikes[i]) + ' bikes on ' + self.timestamp[i].strftime("%Y-%m-%d %H:%M"))



def main():
    file = input("Please enter a file name: ")
    try:
        filename = open(file)
    except:
        print("Please enter a proper file name")

#A file is taken as input and a default error message is shown, if the file entered is invalid. 
#Also the file is read to perform further methods.        
    
    for line in filename:
        if 'id": "20"' in line or 'id": "93"' in line or 'id": "223"' in line or 'id": "307"' in line or 'id": "666"' in line:
            lst1 = line.split(",")
            
            #Name_of_the_station
            stn_name = lst1[2]
            stn_name1 = stn_name.split(":")
            stn_name2 = stn_name1[1]
            stn_name3 = stn_name2.strip(' ""')
            stn.append(stn_name3)
        
            #Bikes_available
            bikes_num = lst1[6]
            bikes_num1 = bikes_num.split(":")
            bikes_num2 = bikes_num1[1]
            bikes_num3 = bikes_num2.strip(' ""')
            bikes_num4 = int(bikes_num3)
            bikes_avail.append(bikes_num4)
           
            #Total_docks
            tot_docks = lst1[3]
            tot_docks1 = tot_docks.split(":")
            tot_docks2 = tot_docks1[1]
            #print(tot_docks2)
            total_docks.append(tot_docks2)
            
            #Docks_in_Service
            dock_serv = lst1[4]
            dock_serv1 = dock_serv.split(":")
            dock_serv2 = dock_serv1[1]
            docks_serv.append(dock_serv2)
            
            #Docks_available
            dock_avail = lst1[5]
            dock_avail1 = dock_avail.split(":")
            dock_avail2 = dock_avail1[1]
            dock_avail3 = dock_avail2.strip(' ""')
            dock_avail4 = int(dock_avail3)
            docks_avail.append(dock_avail4)
            
            #Status
            stat = lst1[8]
            stat1 = stat.split(":")
            stat2 = stat1[1]
            stats.append(stat2)
            
            #Station_ID
            st_id = lst1[0]
            st_id1 = st_id.split(":")
            st_id2 = st_id1[1]
            st_id3 = st_id2.strip(' ""')
            statid.append(st_id3)
            
            #Time_stamp
            tstp = lst1[1]
            tstp1 = tstp.replace('"timestamp": ','').replace("T", " ").replace('"','').strip()
            date_time_obj = datetime.strptime(tstp1, '%Y-%m-%d %H:%M:%S.%f')
            time_stp.append(date_time_obj)
            
            
    #Attributes are passed on to the BikeStation class to perform the functions defined in the BikeStation class. 
    b = BikeStation(statid, time_stp, stn, docks_avail, bikes_avail, stats)
    b.__str__()
    b.present_avail_bikes()
    

if __name__ == "__main__":
    main()
```
